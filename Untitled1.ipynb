```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
match = pd.read_csv('/Users/vikaskumar/Desktop/Jupyter/cric_chatbot/archive/matches.csv')
delivery = pd.read_csv('/Users/vikaskumar/Desktop/Jupyter/cric_chatbot/archive/deliveries.csv')
```

```python
match['year'] = match.apply(lambda x: x.date[-4:], axis=1)

analysis_df = pd.merge(match, delivery, left_on='id', right_on='match_id')

analysis_df_chennai = analysis_df[(analysis_df.city == 'Chennai')]
analysis_df_chennai_csk = analysis_df_chennai[(analysis_df_chennai.team1 == 'Chennai Super Kings') & (analysis_df_chennai.team2 == 'Chennai Super Kings')]

filtered_df = analysis_df_chennai.loc[(analysis_df_chennai['team1'] == 'Chennai Super Kings') | (analysis_df_chennai['team2'] == 'Chennai Super Kings')]
filtered_df
```

```python
filtered_df_match_ids = filtered_df.match_id.unique().tolist()
```

```python
dfs_list = list()
for match_id in filtered_df_match_ids:
    filtered_df_match_id = filtered_df[filtered_df.match_id == match_id]
    filtered_df_match_id_batting_team_csk = filtered_df_match_id[filtered_df_match_id.batting_team == 'Chennai Super Kings']
    filtered_df_match_id_batting_team_csk['dismissal_kind'].fillna(0, inplace=True)
    filtered_df_match_id_batting_team_csk['wickets'] = filtered_df_match_id_batting_team_csk.apply(lambda x: get_wickets_cum_sum(x.dismissal_kind), axis=1)        
    filtered_df_match_id_batting_team_csk['cumulative_runs'] = filtered_df_match_id_batting_team_csk.total_runs.cumsum()
    filtered_df_match_id_batting_team_csk['cumulative_wickets'] = filtered_df_match_id_batting_team_csk.wickets.cumsum()
    filtered_df_match_id_batting_team_csk_ = filtered_df_match_id_batting_team_csk[['id', 'batting_team', 'over', 'cumulative_wickets', 'cumulative_runs']]
    dfs_list.append(filtered_df_match_id_batting_team_csk_)
```

```python
import numpy as np
def get_wickets_cum_sum(dismissal_kind_):
    if dismissal_kind_ != 0:
        return 1
    return 0
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a list of dataframes named 'df_list'
# Each dataframe contains columns ['match_id', 'batting_team', 'over', 'wickets', 'cumulative_runs']

# Concatenate all dataframes into one dataframe
combined_df = pd.concat(dfs_list)

# Group by match_id and over, and calculate the cumulative runs for each over
grouped_df = combined_df.groupby(['id', 'over'])['cumulative_runs'].max().reset_index()

# Calculate the run rate for each over
grouped_df['run_rate'] = grouped_df['cumulative_runs'] / grouped_df['over']

# Plot the run rate over the overs for a sample match
sample_match_id = grouped_df['id'].sample().iloc[0]  # Select a random match for demonstration
sample_match_data = grouped_df[grouped_df['id'] == sample_match_id]

plt.figure(figsize=(10, 6))
plt.plot(sample_match_data['over'], sample_match_data['run_rate'], marker='o', linestyle='-')
plt.title('Run Rate Analysis for Match {}'.format(sample_match_id))
plt.xlabel('Over')
plt.ylabel('Run Rate')
plt.grid(True)
plt.show()

# Compare run rates across matches
plt.figure(figsize=(10, 6))
for match_id, match_data in grouped_df.groupby('id'):
    plt.plot(match_data['over'], match_data['run_rate'], label='Match {}'.format(match_id))

plt.title('Comparison of Run Rates Across Matches')
plt.xlabel('Over')
plt.ylabel('Run Rate')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a list of dataframes named 'df_list'
# Each dataframe contains columns ['match_id', 'batting_team', 'over', 'wickets', 'cumulative_runs']

# Concatenate all dataframes into one dataframe
combined_df = pd.concat(dfs_list)

# Group by match_id and over, and calculate the maximum number of wickets lost in each over
wickets_over_df = combined_df.groupby(['id', 'over'])['cumulative_wickets'].max().reset_index()

# Plotting wickets lost over the overs for a sample match
sample_match_id = wickets_over_df['id'].sample().iloc[0]  # Select a random match for demonstration
sample_match_data = wickets_over_df[wickets_over_df['id'] == sample_match_id]

plt.figure(figsize=(10, 6))
plt.plot(sample_match_data['over'], sample_match_data['cumulative_wickets'], marker='o', linestyle='-')
plt.title('Wickets Lost Over the Overs for Match {}'.format(sample_match_id))
plt.xlabel('Over')
plt.ylabel('Number of Wickets Lost')
plt.grid(True)
plt.show()

# Calculate the average number of wickets lost at different stages of the innings
powerplay_wickets = wickets_over_df[wickets_over_df['over'] <= 6]['cumulative_wickets'].mean()
middle_overs_wickets = wickets_over_df[(wickets_over_df['over'] > 6) & (wickets_over_df['over'] <= 15)]['cumulative_wickets'].mean()
death_overs_wickets = wickets_over_df[wickets_over_df['over'] > 15]['cumulative_wickets'].mean()

print("Average number of wickets lost in Powerplay: {:.2f}".format(powerplay_wickets))
print("Average number of wickets lost in Middle Overs: {:.2f}".format(middle_overs_wickets))
print("Average number of wickets lost in Death Overs: {:.2f}".format(death_overs_wickets))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a list of dataframes named 'df_list'
# Each dataframe contains columns ['match_id', 'batting_team', 'over', 'wickets', 'cumulative_runs']

# Concatenate all dataframes into one dataframe
combined_df = pd.concat(dfs_list)

# Group by match_id and over, and calculate the maximum cumulative runs in each over
cumulative_runs_df = combined_df.groupby(['id', 'over'])['cumulative_runs'].max().reset_index()

# Plotting cumulative runs over the overs for a sample match
sample_match_id = cumulative_runs_df['id'].sample().iloc[0]  # Select a random match for demonstration
sample_match_data = cumulative_runs_df[cumulative_runs_df['id'] == sample_match_id]

plt.figure(figsize=(10, 6))
plt.plot(sample_match_data['over'], sample_match_data['cumulative_runs'], marker='o', linestyle='-')
plt.title('Cumulative Runs Over the Overs for Match {}'.format(sample_match_id))
plt.xlabel('Over')
plt.ylabel('Cumulative Runs')
plt.grid(True)
plt.show()

# Calculate scoring rate (runs per over)
cumulative_runs_df['scoring_rate'] = cumulative_runs_df['cumulative_runs'].diff() / cumulative_runs_df['over'].diff()

# Calculate acceleration rate (change in run rate over consecutive overs)
cumulative_runs_df['acceleration_rate'] = cumulative_runs_df['scoring_rate'].diff()

# Plotting scoring rate over the overs
plt.figure(figsize=(10, 6))
plt.plot(cumulative_runs_df['over'], cumulative_runs_df['acceleration_rate'], marker='o', linestyle='-')
plt.title('Acceleration Rate Over the Overs')
plt.xlabel('Over')
plt.ylabel('Acceleration Rate')
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a list of dataframes named 'df_list'
# Each dataframe contains columns ['match_id', 'batting_team', 'over', 'wickets', 'cumulative_runs']

# Concatenate all dataframes into one dataframe
combined_df = pd.concat(dfs_list)

# Define the phases of the innings (e.g., powerplay, middle overs, death overs)
powerplay_overs = 6
middle_overs = 15
total_overs = 20

# Define a function to classify overs into innings phases
def classify_phase(over):
    if over <= powerplay_overs:
        return 'Powerplay'
    elif over <= middle_overs:
        return 'Middle Overs'
    else:
        return 'Death Overs'

# Apply the function to classify overs into innings phases
combined_df['innings_phase'] = combined_df['over'].apply(classify_phase)

# Group by match_id and innings_phase, and calculate the maximum cumulative runs and wickets lost in each phase
innings_phase_df = combined_df.groupby(['id', 'innings_phase']).agg({'cumulative_runs': 'max', 'cumulative_wickets': 'max'}).reset_index()

# Plotting innings progression across phases
plt.figure(figsize=(12, 6))

# Plot for cumulative runs across phases
plt.subplot(1, 2, 1)
for phase, phase_data in innings_phase_df.groupby('innings_phase'):
    plt.plot(phase_data['cumulative_runs'], label=phase)

plt.title('Innings Progression: Cumulative Runs')
plt.xlabel('Match')
plt.ylabel('Cumulative Runs')
plt.legend()
plt.grid(True)

# Plot for wickets lost acros
```
